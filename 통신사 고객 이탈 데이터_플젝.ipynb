```python
# 필요한 라이브러리 불러오기
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터 로드
file_path = "/content/WA_Fn-UseC_-Telco-Customer-Churn-Updated.csv"  # 파일 경로 확인
data = pd.read_csv(file_path)

# 데이터 크기 및 구조 확인
print("데이터 크기 (행, 열):", data.shape)
print("\n데이터 컬럼 정보:")
print(data.info())
print("\n결측값 확인:")
print(data.isnull().sum())

# 수치형 데이터 통계 분석
print("\n수치형 데이터 통계 요약:")
print(data.describe())

# 범주형 데이터 분포 확인
print("\n범주형 데이터 분포 (Churn):")
print(data['Churn'].value_counts())

# 고객 이탈 비율 확인
print("\n고객 이탈 비율:")
print(data['Churn'].value_counts(normalize=True) * 100)

# TotalCharges 열 변환 확인
print("\nTotalCharges 열의 데이터 타입과 샘플 값 확인:")
print(data['TotalCharges'].dtype)
print(data['TotalCharges'].head(10))

# TotalCharges가 이미 float로 되어 있는 경우 결측값 여부만 확인
print("\nTotalCharges 열 변환 후 결측값 개수:")
print(data['TotalCharges'].isnull().sum())

# 숫자형 열만 선택하여 상관계수 계산
numeric_data = data.select_dtypes(include=['float64', 'int64'])

# 상관계수 계산
print("\n상관계수 (숫자형 변수 간):")
correlation_matrix = numeric_data.corr()
print(correlation_matrix)

# 상관계수 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix (Numeric Data Only)")
plt.show()

# 결측값 처리 (선택적)
# 현재 결측값이 없으므로 처리 필요 없음, 필요 시 아래 코드 참고
# data['TotalCharges'] = data['TotalCharges'].fillna(data['TotalCharges'].mean())

# 이상치 탐지 및 처리
# 수치형 데이터에 대해 Boxplot을 사용하여 확인
plt.figure(figsize=(10, 6))
sns.boxplot(data=numeric_data)
plt.title("Outliers Detection")
plt.show()

# 수치형 데이터 정규화 (선택적)
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
numeric_cols = numeric_data.columns
data[numeric_cols] = scaler.fit_transform(data[numeric_cols])

print("\n정규화된 수치형 데이터 샘플:")
print(data[numeric_cols].head())

# 범주형 변수 인코딩 (Label Encoding)
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
categorical_cols = data.select_dtypes(include=['object']).columns

for col in categorical_cols:
    if col != 'customerID':  # customerID는 인코딩하지 않음
        data[col] = encoder.fit_transform(data[col])

print("\n인코딩 후 데이터 샘플:")
print(data.head())

# 파생 변수 생성
# 예: 월평균 요금 (TotalCharges / tenure, tenure가 0인 경우 NaN으로 처리)
data['MonthlyAverageCharges'] = data['TotalCharges'] / data['tenure'].replace(0, np.nan)

print("\n파생 변수 추가 후 데이터 샘플:")
print(data[['MonthlyAverageCharges']].head())

# 불필요한 열 제거
if 'customerID' in data.columns:
    data = data.drop(columns=['customerID'])

print("\n불필요한 열 제거 후 데이터 크기:", data.shape)
```

```python
#######데이터 탐색

# 데이터 구조 확인
print(data.info())

# 기본 통계량 확인
print(data.describe())

# 결측치 확인
print(data.isnull().sum())

# 주요 특성 탐색: Churn 분포 확인
print(data['Churn'].value_counts(normalize=True))

# 숫자형 데이터 분포 확인
print(data[['tenure', 'MonthlyCharges', 'TotalCharges']].describe())
```

```python
########데이터전처리

# 결측치 처리
# TotalCharges의 결측치는 tenure가 0인 고객에게 발생한 것으로 보임 -> 0으로 대체
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'].fillna(0, inplace=True)

# 불필요한 컬럼 제거
data = data.drop(['customerID'], axis=1)

# 범주형 변수 인코딩
data = pd.get_dummies(data, drop_first=True)

# 스케일링 (연속형 변수)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(data[['tenure', 'MonthlyCharges', 'TotalCharges']])

# 데이터 구조 및 정제 결과 확인
print(data.info())
```

```python
######데이터 전처리


# TotalCharges: 결측치 처리
# TotalCharges는 tenure가 0인 고객에게 결측치 발생 -> 0으로 대체
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data['TotalCharges'] = data['TotalCharges'].fillna(0)

# 불필요한 컬럼 제거
if 'customerID' in data.columns:
    data = data.drop(['customerID'], axis=1)

# 범주형 변수 인코딩
data = pd.get_dummies(data, drop_first=True)

# 스케일링 (연속형 변수)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data[['tenure', 'MonthlyCharges', 'TotalCharges']] = scaler.fit_transform(data[['tenure', 'MonthlyCharges', 'TotalCharges']])

# 데이터 구조 및 정제 결과 확인
print(data.info())
```

1. 고객 이탈 비율 시각화 (Pie Chart & Bar Chart)


```python
# 고객 이탈 비율 시각화
churn_counts = data['Churn'].value_counts()
churn_labels = ['No Churn', 'Churn']

# Pie Chart
plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=churn_labels, autopct='%1.1f%%', startangle=90, counterclock=False, colors=['skyblue', 'orange'])
plt.title('Customer Churn Proportion')
plt.show()

# Bar Chart
plt.figure(figsize=(6, 4))
churn_counts.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Customer Churn Count')
plt.xlabel('Churn')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=churn_labels, rotation=0)
plt.show()
```

2. 범주형 변수와 이탈 간 관계 (Stacked Bar Chart)


```python
# 범주형 변수와 이탈 관계: Stacked Bar Chart
categorical_vars = ['Contract', 'InternetService', 'PaymentMethod']
for var in categorical_vars:
    churn_ct = pd.crosstab(data[var], data['Churn'], normalize='index') * 100
    churn_ct.plot(kind='bar', stacked=True, figsize=(8, 4), color=['skyblue', 'orange'])
    plt.title(f'Churn by {var}')
    plt.xlabel(var)
    plt.ylabel('Percentage')
    plt.legend(['No Churn', 'Churn'], title='Churn', loc='upper right')
    plt.show()
```

3. 연속형 변수의 분포와 이탈 (Box Plot)


```python
# Box Plot for Continuous Variables
plt.figure(figsize=(12, 6))
for i, col in enumerate(['tenure', 'MonthlyCharges', 'TotalCharges'], 1):
    plt.subplot(1, 3, i)
    sns.boxplot(data=data, x='Churn', y=col, palette='Set2')
    plt.title(f'{col} vs Churn')
    plt.xlabel('Churn')
    plt.ylabel(col)
plt.tight_layout()
plt.show()
```

4. Tenure에 따른 이탈율 변화 (Line Plot)


```python
# Tenure에 따른 이탈율 계산
tenure_churn = data.groupby('tenure')['Churn'].mean()

# Line Plot
plt.figure(figsize=(8, 5))
tenure_churn.plot(kind='line', marker='o', color='orange')
plt.title('Churn Rate by Tenure')
plt.xlabel('Tenure (Months)')
plt.ylabel('Churn Rate')
plt.grid()
plt.show()
```

5. 상관관계 분석 (Heatmap)


```python
# Heatmap for Correlation
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True, fmt='.2f', cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```
